````python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv('./data/college_completion.csv')
print("Dimensions: ", df.shape)
print("Observations: ", df.shape[0])
print("Variables: ", df.columns)
print(df.head())
````

````python
ctab = pd.crosstab(df['control'], df['level'])
print(ctab)
````

````python
sns.histplot(df['grad_100_value'])
plt.title("Histogram of Graduation Rate")
plt.show()

sns.kdeplot(df['grad_100_value'])
plt.title("KDE of Graduation Rate")
plt.show()

sns.boxplot(x=df['grad_100_value'])
plt.title("Boxplot of Graduation Rate")
plt.show()

print(df['grad_100_value'].describe())
````

````python
sns.kdeplot(data=df, x='grad_100_value', hue='control')
plt.title("Graduation Rate by Control")
plt.show()

sns.kdeplot(data=df, x='grad_100_value', hue='level')
plt.title("Graduation Rate by Level")
plt.show()
````

Public and private for profit has the lowest graduation rate, with public being slightly higher. Private non profit has significantly higher graduation rate. 4 year has significantly higher graduation rate than 2 year.

````python
print(df.groupby('control')['grad_100_value'].describe())
print(df.groupby('level')['grad_100_value'].describe())
````

4 year private non profit has the best graduation rate because it has the highest mean/median.

````python
df['levelXcontrol'] = df['level'] + ', ' + df['control']

sns.kdeplot(data=df, x='grad_100_value', hue='levelXcontrol')
plt.title("Graduation Rate by Level and Control")
plt.show()
````

````python
sns.kdeplot(data=df, x='aid_value')
plt.title("KDE of Aid Value")
plt.show()

sns.kdeplot(data=df, x='aid_value', hue='levelXcontrol')
plt.title("Aid Value by Level and Control")
plt.show()

print(df.groupby(['level', 'control'])['aid_value'].describe())
````

All institutions have a graph that's heavily skewed right, except for 4 year private non profit which is much more centered and less right skewed compared to the other institutions.

````python
sns.scatterplot(data=df, x='aid_value', y='grad_100_value')
plt.title("Graduation Rate vs Aid")
plt.show()
````

There is a slight positive linear correlation where higher aid value results in higher graduation rate. However, this trend doesn't seem to apply for aid values lower than 10000 as the graduation rate is high regardless.

Private non profit gives the most aid by a large margin. Private for profit gives the least aid. Public only gives slightly more aid than private for profit. For 4 year private non profit institutions, aid seems to increase graduation rate.

3. The **hazard rate** gives the instantaneous risk of an event occurring, given that it hasn't already happened. It is defined as
$$
h(t) = \dfrac{f(t)}{1-F(t)}
$$
where $F(t)$ is the distribution of arrival times and $f(t)=F'(t)$ is the density of arrival times. Roughly, conditional on $T \ge t$, what is the probability that $T=t$? The cumulative hazard is given by
$$
H(t) = \int_0^t h(z) dz.
$$
The following function computes the hazard rate and cumulative hazard using something called the Nelson-Ahlen estimator. Run your data through it. Interpret the plots for the hazard rate and cumulative hazard rate. What patterns do you see?

```
def hazard(df):
    """ Compute cumulative hazard rate using Nelson-Ahlen estimator. """
    at_risk = df.shape[0]
    arrivals = df['survival'].sort_values().dropna().tolist()
    hazard_rate = []
    times = []
    for t in arrivals:
        select = (df['survival']==t)*(df['Overall Survival Status']=='Deceased')
        deaths_t = df[select].shape[0]
        if deaths_t > 0:
            hazard_rate.append(deaths_t/at_risk)
            times.append(t)
            at_risk -= deaths_t
    cumulative_hazard = np.cumsum(hazard_rate)
    return cumulative_hazard, times, hazard_rate
```
